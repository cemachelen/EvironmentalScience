**Q2. What is the total cost of the fruit if 6 bananas at 0.2 each are added to the shopping list?**


```python
# Assign a value to each variable
apples = 3
oranges = 6
bananas = 6.0 

price_per_banana = 0.2
price_per_apple = 0.5  
price_per_orange = 0.3  

# Calculate the total cost
cost_of_fruit = (price_per_apple * apples) + (price_per_orange * oranges)+(price_per_banana * bananas)

# Print answer
print_format_cost = 'Total cost of the fruit = '
print (print_format_cost, cost_of_fruit)
```

**Q4.  Use numpy commands to calculate the maximum and minimum temperatures in the temperature array.** 

```python
# Import Modules (loads a lot of useful functions)
import numpy as np
import matplotlib.pyplot as plt
from pylab import rcParams
import pandas as pd
import  scipy.io.netcdf as netcdf
import os
# This line allows the plots to show in the notebook
%matplotlib inline
# This sets the size of the plot
rcParams['figure.figsize'] = (8,6)
# Define a data file 
csv_file_name = "Leeds_Temperature_Data_20151101_20151130_Simple.csv"
# Read that data file
data = pd.read_csv(csv_file_name)
# Extract columns Date and Temp
timestamp = data['Date_Time']
temperature = data['Temp']

# Print a string and a variable
print('Tmin = ', temperature.min())
print('Tmax = ', temperature.max())

# OR
print('Or formatted nicely:')

# Print a string and variable cobined into a sentence
print('Tmin = '+str(temperature.min()))
print('Tmax = '+str(temperature.max()))
```

<hr>

**Optional advanced Q1:  The minimum temperature is: 0.5,0.3,1.7,3.2,5.9,8.8,11.1,10.9,8.9,6.1,3.2,0.9  Replot the previous plot, but also plot the min_temperature on the same plot.**



```python
# Temperature arrays
max_temperature = [5.5,5.6,8.1,10.7,14.3,16.9,19.1,18.7,15.9,12.1,8.3,5.8]
min_temperature = [0.5, 0.3, 1.7, 3.2, 5.9, 8.8, 11.1, 10.9, 8.9, 6.1, 3.2, 0.9]
# Turn list into array
max_temperature = np.array(max_temperature)
# List months
month = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sept',
         'Oct', 'Nov', 'Dec']
# formatter for nicely labeling months
xticks_position = range(0,len(month),1)

# plot max and min temps
plt.plot(max_temperature)
plt.plot(min_temperature)
# Format the month lables
plt.xticks(xticks_position,month)
# Lable axis and title
plt.title('Q 1 Plot',fontsize=14)
plt.xlabel("Month")
plt.ylabel("Temperature, Degrees $^o$C")
# Add a ledgend
plt.legend(['Tmax','Tmin'])
plt.show()
```

<hr>
**Optional advanced Q2:  Calculate the mean (average) temperature from the maximum and minimum temperature arrays, and plot this on the graph**

```python
# Plot max and min temperature
plt.plot(max_temperature, color = 'orange')
plt.plot(min_temperature, color = 'cyan')
# Make the lists into numpy arrays so we can mean etc
min_temperature = np.array(min_temperature)
max_temperature = np.array(max_temperature)
# plot a horizontal line at the mean max and min temperatures
plt.axhline(max_temperature.mean(), color='red')
plt.axhline(min_temperature.mean(), color='blue')
# Formtat months
plt.xticks(xticks_position,month)
# Add title, axis labels and legend
plt.title('Q 2 Plot',fontsize=14)
plt.xlabel("Month")
plt.ylabel("Temperature, Degrees $^o$C")
plt.legend(['Tmax','Tmin','Maxmean','Minmean'])
plt.show()
```

**Optional advanced Q3: Re-plot the plot, but using the date / time from the timestamp array as xtick marks.**

```python
# You can use pandas date time and resample and plot
# format column date_time to a python date-time
data['Date_Time'] = pd.to_datetime(data['Date_Time']) 
df = data.set_index('Date_Time')  # index by date-time
hourly_mean = df.resample('H').mean()  # Resample hourly mean
# Plot the new hourly mean
plt.plot(hourly_mean['2015-11-15':'2015-12-01']) # plot date range 15/11 to 01/12
plt.ylabel(r'Temp ($^oC$)',fontsize=16) # r flag allows for tex symbols
plt.xlabel('Date',fontsize=16)
plt.title('Resampled Hourly Mean Temperatures',fontsize=20)
plt.xticks(rotation='vertical') # Rotate xticks
plt.show()
```

```python
# The non - pandas way!
# 24 hours 7 days a week
frequency = 24 * 7
# Genertate the positions of the x ticks
position = (range(0,len(timestamp),frequency))
# Generate the labels
lables = (timestamp[range(0,len(timestamp),frequency)])
# Make them veritcal
plt.xticks(position, lables , rotation='vertical')
# plot and label
plt.plot(temperature)
plt.xlabel("Time / Date")
plt.ylabel("Temperature, Degrees C")
plt.show()
```

**Q5. Extend the map to cover the UK, Spain and Western Europe**



```python
# Plot a map centered on the UK.
# Map coordinates are: 15W to 10E, 35N to 65N.
# A nice thing to do is asing east west , north south values
E, W, N, S = [10, -15, 65, 35]
# These then can go in place of numbers in your map settings 
# Allows you to see more easily western boundary, Southern boundary
# E.g. change 35 to 30 and see what happens
map = Basemap(projection='cyl',lat_0=0,lon_0=0, llcrnrlon=W, 
              llcrnrlat=S, urcrnrlon=E, urcrnrlat=N)
map.drawcoastlines()
map.drawcountries()
map.fillcontinents(color='coral')
# Show the map
plt.show()
```

**Advanced Q5: Add site_name lables to the map so its clear which site is which.**

```python
# List sites and corresponding co-ordinates and temps
temperature = [9,5,7,8,8]
site_name = ["Edinburgh","Leeds","London","Inverness","Exeter"]
latitudes = [55.9,53.8,51.5,57.5,50.7]
longitudes = [-3.2,-1.5,-0.1,-4.2,-3.5]
# Set up the map as before
E, W, N, S = [10, -15, 65, 35]
map = Basemap(projection='cyl',lat_0=0,lon_0=0, llcrnrlon=W, 
              llcrnrlat=S, urcrnrlon=E, urcrnrlat=N)
map.bluemarble()
x,y = map(longitudes, latitudes)
map.scatter(x, y, s=temperature,marker="o")
# for each site and correspoding point plot
for label, xpt, ypt in zip(site_name, x, y):
    # plot the text of the site_name in bold allinged a certain way
    # colour k = black
    plt.text(xpt, ypt, label,fontsize=12,fontweight='bold',
             ha='left',va='bottom',color='k')
plt.show()
```

**Advanced Q6. Its hard to tell the different marker sizes apart. Can you replot the map using
colour to show the temperature?**


```python
# Measured temperature
temperature = [9,5,7,8,8]
site_name = ["Edinburgh","Leeds","London","Inverness","Exeter"]
latitudes = [55.9,53.8,51.5,57.5,50.7]
longitudes = [-3.2,-1.5,-0.1,-4.2,-3.5]
E, W, N, S = [10, -15, 65, 35]
map = Basemap(projection='cyl',lat_0=0,lon_0=0, llcrnrlon=W, 
              llcrnrlat=S, urcrnrlon=E, urcrnrlat=N)
map.bluemarble()
x,y = map(longitudes, latitudes)
# you can use a scatter plot
# points x and y with size (s) and colour (s) correspondind to temperature
# on a colour map (cmap) from blue to white to red
map.scatter(x, y , s=temperature,c=temperature,marker="o",cmap=plt.cm.bwr)
# add the labels like before
for label, xpt, ypt in zip(site_name, x, y):
    plt.text(xpt, ypt, label,fontsize=12,fontweight='bold',
             ha='left',va='bottom',color='k')
plt.show()
```
